2. Manually derive biologically sensible parameter values to define transition matrix ​A​2 and emission matrix ​E2​. You may assume that hydrophobic residues are more common in domains and less abundant in linker regions, while charged and polar amino acids are more common in linker regions and less common in domains. Assume linker regions exclusively occur between two domains. 

For reference, the A1 and E1 priors used in the previous part are given by:

|   | B   | L   | D   | E   |
|---|-----|-----|-----|-----|
| B | 0   | 0.5 | 0.5 | 0   |
| L | 0   | 0.7 | 0.2 | 0.1 |
| D | 0   | 0.2 | 0.7 | 0.1 |
| E | 0   | 0   | 0   | 0   |

|   | H   | P   | C   |
|---|-----|-----|-----|
| L | 0.5 | 0   | 0.5 |
| D | 0   | 0.5 | 0.5 |

We can observe that these matrices are not biologically sensible, at least not according to the biological prior knowledge given above. If hydrophobic residues are more common in domains and less abundant in linker regions, the emission probability of H should be higher given D than given L. Furthermore, if polar and charged amino acids are more common in linker regions than in domains, the emission probabilities of P and C should be higher given L than given D. A more sensible prior for the emission probabilities could thus be the following:

|   | H   | P   | C   |
|---|-----|-----|-----|
| L | 0.1 | 0.45 | 0.45|
| D | 0.5 | 0.25 | 0.25|

In this matrix, the ordering of the relationships between the probabilities coheres with the given biological knowledge.

If linker regions exclusively appear between two domains, we should modify the transition matrix so that the begin state can only enter into a domain and the end state can only be entered from a domain. Thus, a more biologically sensible prior for the transition matrix could be

|   | B   | L   | D   | E    |
|---|-----|-----|-----|------|
| B | 0   | 0   | 1    | 0   |
| L | 0   | 0.75| 0.25 | 0   |
| D | 0   | 0.2 | 0.7  | 0.1 |
| E | 0   | 0   | 0    | 0   |

We will now train using the Baum-Welch algorithm using these parameters.

```python
!python3 hmm.py -v -o output/output_A2_E2 baumwelch input/training_sequences.fasta input/A2_prior.tsv input/E2_prior.tsv
```

It is noteworthy that convergence is much faster, although the log-likelihood only differs by about 10 units.
